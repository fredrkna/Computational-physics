```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
arts = 100
time = 1000
n = 1000

treshold = np.zeros(time)

for a in range(n):
    evolution = np.zeros([time, arts])
    fitness = np.zeros(arts)


    for i in range(arts):
        fitness[i] = np.random.uniform()

    for i in range(time):
        r = np.argmin(fitness)
        treshold[i] += np.partition(fitness, 5)[5]
        evolution[i][r] = 1
        fitness[r] = np.random.uniform()
        if r != 0 and r != arts-1:
            fitness[r+1] = np.random.uniform()
            fitness[r-1] = np.random.uniform()
            evolution[i][r+1] = 1
            evolution[i][r-1] = 1
        elif r == 0:
            fitness[-1] = np.random.uniform()
            fitness[r+1] = np.random.uniform()
            evolution[i][-1] = 1
            evolution[i][r+1] = 1
        else:
            fitness[r-1] = np.random.uniform()
            fitness[0] = np.random.uniform()
            evolution[i][r-1] = 1
            evolution[i][0] = 1

treshold = treshold/n
            
plt.figure()
plt.plot(treshold)
plt.xlabel("Time",fontsize=18)
plt.ylabel("<Treshold>",fontsize=18)
plt.show()
```

```python
from matplotlib import pyplot as plt

# Fixing random state for reproducibility
#np.random.seed(19680801)


X = evolution

fig, ax = plt.subplots(figsize=(100, 500))
ax.imshow(X, interpolation='nearest')

numrows, numcols = X.shape

def format_coord(x, y):
    col = int(x + 0.5)
    row = int(y + 0.5)
    if col >= 0 and col < numcols and row >= 0 and row < numrows:
        z = X[row, col]
        return 'x=%1.4f, y=%1.4f, z=%1.4f' % (x, y, z)
    else:
        return 'x=%1.4f, y=%1.4f' % (x, y)

ax.format_coord =  format_coord
plt.show()
```
